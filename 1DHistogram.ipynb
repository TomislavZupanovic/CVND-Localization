# Initialize environment

First we initialize 1D environment, measurements and motions for 1D Filter using colors for environment and measurements and numbers for motions. Object will be "sensing" color in its surrounding and changing distribution (gaining information) where it is likely located, then it will move based on motion, therefore moving location distribution in motion direction (losing information). We repeat this process in cycle.

<img src='images/sense_move.png' width=50% height=50% />

```python
from source.histogram_1D import Filter1D
```

```python
# 1D world of colors
environment = ['blue','blue','yellow','blue','yellow','yellow','blue','blue','blue','blue']
# Color sensor measurements from object
measurements = ['blue','yellow','blue','yellow','yellow']
# Taken motions after measurements
# We expect to be at 3rd position from end after these motions
motions = [1, 1, 1, 1, 2]
```

**Initialize 1D Filter and display prior distribution:**

```python
filter_1d = Filter1D(environment)
filter_1d.display_map()
```

**Cycle through all the measurements and motions then display the posterior probability. Object is most likely at position 7, as expected:**

```python
filter_1d.cycle(motions, measurements)
filter_1d.display_map()
```
