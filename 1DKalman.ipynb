## 1D Kalman Filter

Just like 1D Histogram Filter, object in the environment is sensing and updating its location probability by measuring surrounding, then predicting it's next location probability by using motion Gaussian. Starting from initial very uncertain location Gaussian to a more and more certain Gaussian by performing this update and predict actions.

**Initialize 1D environment**

```python
from source.kalman_1D import Kalman1D
```

```python
# Mean values of measurement and motion Gaussians
measurements = [5., 6., 7., 9., 10., 12., 13., 16., 17.]
motions = [1., 1., 2., 1., 2., 1., 3., 1.]
```

**Kalman Filter and initial uncertainty**

Initial estimate is set to the location 0, and the variance is extremely large; this is a state of high confusion so we don't know where object actually is. We have variance in both measurement and motion, since neither of those readings are perfect.

```python
# 0 = initial mean, 1000 = initial variance
# 3 = measure variance, 2 = motion variance
kalman = Kalman1D(0, 1000, 3, 2)
# Plot initial Gaussian distribution
kalman.display_distribution()
```

**Cycle the update and predict operations in the environment**

After all defined measurements and motions with its uncertainties, we expect object to be around 17 with most probability and less probability anywhere else.

```python
kalman.cycle(motions, measurements)
kalman.display_distribution()
```
